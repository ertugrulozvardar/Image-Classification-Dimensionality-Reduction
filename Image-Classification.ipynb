```python
import mnist_reader
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
#Loading the dataset (training and test sets separately)
x_train, y_train = mnist_reader.load_mnist('data', kind='train')
x_test, y_test = mnist_reader.load_mnist('data', kind='t10k')
```

```python
# Visualizing the first 8 images.
for i in range(8):
    plt.figure()
    plt.imshow(x_train[i].reshape(28,28), cmap='gray')
```

```python
# Normalizing pixels into range of 0 and 1
scaler = MinMaxScaler(feature_range = (0,1))
scaled_x_train = scaler.fit_transform(x_train)
scaled_x_test = scaler.transform(x_test)
```

```python
# Visualizing the first 8 images.
for i in range(8):
    plt.figure()
    plt.imshow(scaled_x_train[i].reshape(28,28), cmap='gray')
```

```python
# Initializing KNeighbors Classifier with its parameters.
from sklearn.neighbors import KNeighborsClassifier
##  p : int, default=2
    #Power parameter for the Minkowski metric. When p = 1, this is
    #equivalent to using manhattan_distance (l1), and euclidean_distance
    #(l2) for p = 2. For arbitrary p, minkowski_distance (l_p) is used.
knn=KNeighborsClassifier(n_neighbors=3, p=2, metric='minkowski') 

knn.fit(scaled_x_train, y_train) # we only use the training data and their lables to fit the model !
```

```python
# Predicting the classes
knn_prediction=knn.predict(scaled_x_test) # we predict the class labels of the instances in test set
```

```python
# Visualizing the images with their predicted and true labels. -For KNN
for i in range(13):
    plt.figure()
    plt.title(f"True label:{y_test[i]}, Predicted label:{knn_prediction[i]}")
    plt.imshow(scaled_x_test[i].reshape(28,28), cmap='gray')
```

```python
# Evaluating the performance of the KNeighbors model with classification accuracy
from sklearn.metrics import accuracy_score
knn_accuracy = accuracy_score(knn_prediction, y_test)
```

```python
# Visualize the confusion matrix of the predicted values by K Neighbors Classifier
import seaborn as sns
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, knn_prediction)
plt.figure(figsize=(6,6))
sns.heatmap(conf_matrix, annot=True)
```

```python
# Initializing Perceptron Classifier with its parameters.
from sklearn.linear_model import Perceptron
# Create a perceptron object with the parameters: 40 iterations (epochs) over the data, and a learning rate of 0.1
ppn = Perceptron(max_iter=40, eta0=0.1, random_state=0)

# Train the perceptron
ppn.fit(scaled_x_train, y_train)
```

```python
# Predicting the classes.
# Apply the trained perceptron on the X data to make predicts for the y test data.
ppn_prediction = ppn.predict(scaled_x_test)
```

```python
# Visualizing the images with their predicted and true labels. -For Perceptron.
for i in range(13):
    plt.figure()
    plt.title(f"True label:{y_test[i]}, Predicted label:{ppn_prediction[i]}")
    plt.imshow(scaled_x_test[i].reshape(28,28), cmap='gray')
```

```python
# Evaluating the performance of the Perceptron model with classification accuracy.
from sklearn.metrics import accuracy_score
ppn_accuracy = accuracy_score(ppn_prediction, y_test)
```

```python
# Visualize the confusion matrix of the predicted values by Perceptron Classifier
conf_matrix = confusion_matrix(y_test, ppn_prediction)
plt.figure(figsize=(6,6))
sns.heatmap(conf_matrix, annot=True)
```

```python
# Dimensionality Reduction
# Finding the number of components required to explain P% of the variance of the dataset where P approximates to 25, 50, 75, 95 for each

from sklearn.decomposition import PCA
num_of_components_p = dict()
p_values = [0.25, 0.50, 0.75, 0.95]
pca_list = []
for x in p_values:
    pca = PCA(n_components = x)
    pca.fit(scaled_x_train)
    pca_list.append(pca)
    num_of_components_p[x] = len(pca.components_)
```

```python
import pandas as pd
# Showing the number of components which respect to P percentages.
pd.DataFrame(num_of_components_p, index = ["Number of Components"])
```

```python
# For each P, repeat the training and evaluation parts with the data 
# reduced its dimensionality by PCA with “n_components” found in the previous step. - P = 25% - KNN
X_transformed = pca_list[0].fit_transform(scaled_x_train)

knn_pca_25 = KNeighborsClassifier(n_neighbors=3, p=2, metric='minkowski') 
knn_pca_25.fit(X_transformed, y_train)
newdata = scaled_x_test

newdata_transformed = pca_list[0].transform(newdata)

pred_labels = knn_pca_25.predict(newdata_transformed)
# Evaluating the performance of the RFC model with classification accuracy.
acc_score_knn_25 = accuracy_score(pred_labels, y_test)
```
